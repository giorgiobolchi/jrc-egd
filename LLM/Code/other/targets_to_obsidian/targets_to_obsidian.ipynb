```python
import os
import pandas as pd
import openpyxl
```

```python
# Path to the CSV file and output directory
date = '20250312'
file_path = '/Users/giorgiobolchi2/Documents/GitHub/jrc-egd/LLM/Code/targets_to_obsidian/data/v3/target_data_v3.xlsx'
output_dir = f'/Users/giorgiobolchi2/Documents/GitHub/jrc-egd/LLM/Code/targets_to_obsidian/EGD_map_auto/{date}'

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

```

```python
df = pd.read_excel(file_path, engine='openpyxl')  # Read the .xlsx file

    
# Print first 5 rows to verify the data 
print(df.head())
print('Total rows:',len(df))
```

```python
# Define the subdirectory names for each 'TA' prefix
subdirectories_names = {
    'TA1': 'TA1_Climate ambition',
    'TA2': 'TA2_Clean, affordable and secure energy',
    'TA3': 'TA3_Industrial strategy for a clean and circular economy',
    'TA4': 'TA4_Sustainable and smart mobility',
    'TA5': 'TA5_Greening the Common Agricultural Policy - Farm to Fork Strategy',
    'TA6': 'TA6_Preserving and protecting biodiversity',
    'TA7': 'TA7_Towards a zero-pollution ambition for a toxic free environment',
}

for _, row in df.iterrows():
    
    # Extract details from the columns
    target_code = row.get('target_code', 'Untitled')
    target_content = row.get('target_content', 'No content provided')
    target_assessment = row.get('target_assessment', 'No assessment')
    thematic_area = row.get('thematic_area', 'No thematic area')
    assessment_colour = row.get('assessment_colour', 'No tags')

    # Convert target_code to string if it is not already
    if not isinstance(target_code, str):
        target_code = str(target_code)

    # Extract the 'TA' prefix from the target_code
    ta_prefix = target_code.split('.')[0]

    # Create the subdirectory name based on the 'TA' prefix
    subdirectory_name = subdirectories_names.get(ta_prefix, f"{ta_prefix}_Unknown")

    # Create the subdirectory if it doesn't exist
    subdirectory_path = os.path.join(output_dir, subdirectory_name)
    os.makedirs(subdirectory_path, exist_ok=True)

    # Create the markdown content
    markdown_content = f"## {target_content}\n\nThematic area: #{thematic_area}\n\nAssessment: {target_assessment}\n\nColour: {assessment_colour}"  

    # Save to a markdown file
    file_path = os.path.join(subdirectory_path, f"{target_code}.md")
    with open(file_path, 'w', encoding='utf-8') as md_file:
        md_file.write(markdown_content)

    # Print statement to check each iteration
    print(f"Created file: {file_path}")

print("Notes created.")
```
